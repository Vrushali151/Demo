```python
files=['Copper_2023.pdf','Copper_2023.csv','Copper_2022.pdf','Copper_2022.csv','Copper_2024.pdf','Al_2022.pdf','Al_2023.csv','Al_2024.pdf'
       ,'Al_2024.csv','Iron_2023.pdf','Iron_2022.csv','Iron_2024.pdf','Silver_2023.pdf','Silver_2023.csv','Silver_2024.csv',
       'Silver_2024.pdf','Silver_2022.csv']
Copper_csv_2022= []
Copper_pdf_2022=[]
Copper_csv_2023= []
Copper_pdf_2023=[]
Copper_csv_2024= []
Copper_pdf_2024=[]
Al_csv_2022= []
Al_pdf_2022=[]
Al_csv_2023= []
Al_pdf_2023=[]
Al_csv_2024= []
Al_pdf_2024=[]
Iron_csv_2022= []
Iron_pdf_2022=[]
Iron_csv_2023= []
Iron_pdf_2023=[]
Iron_csv_2024= []
Iron_pdf_2024=[]
Silver_csv_2022= []
Silver_pdf_2022=[]
Silver_csv_2023= []
Silver_pdf_2023=[]
Silver_csv_2024= []
Silver_pdf_2024=[]

for filename in files:
    if filename.startswith("Copper") :#Code  for To check Copper file
         
        if filename.endswith(".pdf"):
                if '2022' in filename:
                     Copper_pdf_2022.append(filename)
                    
                elif '2023'in filename:
                     Copper_pdf_2023.append(filename)
                    
                elif '2024'in filename:
                     Copper_pdf_2024.append(filename)
                     

        
        elif filename.endswith(".csv"):
                if '2022' in filename:
                     Copper_csv_2022.append(filename)
                     
                elif '2023'in filename:
                     Copper_csv_2023.append(filename)
                     
                elif '2024'in filename:
                     Copper_csv_2024.append(filename)
                     
    
    elif filename.startswith("Al") :#Code  for To check Al file
         
        if filename.endswith(".pdf"):
                if '2022' in filename:
                     Al_pdf_2022.append(filename)
                     
                elif '2023'in filename:
                     Al_pdf_2023.append(filename)
                      
                elif '2024'in filename:
                     Al_pdf_2024.append(filename)
                     

        
        elif filename.endswith(".csv"):
                if '2022' in filename:
                     Al_csv_2022.append(filename)
                     
                elif '2023'in filename:
                     Al_csv_2023.append(filename)
                    
                elif '2024'in filename:
                     Al_csv_2024.append(filename)
                     
    
    elif filename.startswith("Iron") :#Code  for To check Iron file
         
        if filename.endswith(".pdf"):
                if '2022' in filename:
                     Iron_pdf_2022.append(filename)
                     
                elif '2023'in filename:
                     Iron_pdf_2023.append(filename)
                      
                elif '2024'in filename:
                     Iron_pdf_2024.append(filename)
                    

        
        elif filename.endswith(".csv"):
                if '2022' in filename:
                     Iron_csv_2022.append(filename)
                     
                elif '2023'in filename:
                     Iron_csv_2023.append(filename)
                    
                elif '2024'in filename:
                     Iron_csv_2024.append(filename)
                    
    
            
    elif filename.startswith("Silver") :#Code  for To check Silver file
         
        if filename.endswith(".pdf"):
                if '2022' in filename:
                     Silver_pdf_2022.append(filename)
                     
                elif '2023'in filename:
                     Silver_pdf_2023.append(filename)
                    
                elif '2024'in filename:
                     Silver_pdf_2024.append(filename)
                     

        
        elif filename.endswith(".csv"):
                if '2022' in filename:
                     Silver_csv_2022.append(filename)
                    
                elif '2023'in filename:
                     Silver_csv_2023.append(filename)
                     
                elif '2024'in filename:
                     Silver_csv_2024.append(filename)
                     
print("Copper Files :-")       
print(Copper_pdf_2022) 
print(Copper_pdf_2023)                  
print(Copper_pdf_2024)   
print(Copper_csv_2022)
print(Copper_csv_2023)
print(Copper_csv_2024)
print("AL Files :-")       
print(Al_pdf_2022) 
print(Al_pdf_2023)                  
print(Al_pdf_2024)   
print(Al_csv_2022)
print(Al_csv_2023)
print(Al_csv_2024)
print("Iron Files :-")       
print(Iron_pdf_2022) 
print(Iron_pdf_2023)                  
print(Iron_pdf_2024)   
print(Iron_csv_2022)
print(Iron_csv_2023)
print(Iron_csv_2024)
print("Silver Files :-")       
print(Silver_pdf_2022) 
print(Silver_pdf_2023)                  
print(Silver_pdf_2024)   
print(Silver_csv_2022)
print(Silver_csv_2023)
print(Silver_csv_2024)
```
